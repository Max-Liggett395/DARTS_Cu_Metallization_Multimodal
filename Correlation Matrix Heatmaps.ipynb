```python
#importing libraries needed for heatmaps
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import colormaps
from matplotlib.colors import ListedColormap
```

```python
#creating initial printing heatmap

#import spreadsheet into file
CuData = pd.read_excel(*file path*)

#columns for printing parameters
printing_cols = [5, 6, 7, 8, 9, 14, 16, 22, 24]
Cu_printing = CuData.iloc[:, printing_cols]

#print the printing parameters taken from the master data sheet
"""
print('Printing Parameters: \n')
print(Cu_printing)
"""

#loop through dataframe and create dummy variables for each column
#while(val < len(CuAg_printing)
printing_removed = Cu_printing.drop(index = 0)
printing_complete = printing_removed.dropna().astype('float')

#print('Printing Complete: \n')
#print(printing_complete)

#creating correlation used for heatmap
correlation_printing = printing_complete.corr(method='pearson')

plt.figure(figsize=(30, 20)) #changing size of heatmap
plt.title("Printing Initial Heatmap", fontsize=60) #creating title for heatmap

p1 = sns.heatmap(correlation_printing, cmap = 'summer', center = 0.75, linewidths = 7, annot = True, annot_kws={"size":25}, square = True,
                    xticklabels = True, yticklabels = True, linecolor='black')
p1.set_xticklabels(p1.get_xticklabels(), rotation=20, fontsize=20)
p1.set_yticklabels(p1.get_yticklabels(), rotation=0, fontsize=25)

#saving heatmap as a png image
printing_path = *file path*
plt.savefig(printing_path)
```

```python
#making a printing heatmap with no empty spaces

#removing the unchanging rows from the heatmap
changing_vals = [2, 3, 6, 7, 8] 
printing_final = printing_complete.iloc[:, changing_vals]

#creating correlation used for heatmap
correlation_final = printing_final.corr(method='pearson')

plt.figure(figsize=(15, 15)) #changing size of heatmap
plt.title("Printing Final Heatmap", fontsize=40) #creating title for heatmap

#creating heatmap with seaborn
p1_final = sns.heatmap(correlation_final, cmap = 'Reds', linewidths = 3, annot = True, annot_kws={"size":25}, square = True)
p1_final.set_xticklabels(p1_final.get_xticklabels(), fontsize=13)
p1_final.set_yticklabels(p1_final.get_yticklabels(), fontsize=13)

#saving heatmap as a png image
printing_path = *file path*
plt.savefig(printing_path)
```

```python
#creating initial sintering heatmap

#sintering parameters
sintering_cols = [10, 11, 12, 13, 15, 20, 25] #columns with sintering metrics
Cu_sintering = CuData.iloc[:, sintering_cols] #creating DataFrame with only sintering

#removing the first row from the dataframe
sintering_removed = Cu_sintering.drop(index = 0)

#removing NaN, 'unknown', and non-numerical values
sintering_complete = sintering_removed.replace('unknown', np.nan).infer_objects(copy=False)
sintering_complete = sintering_complete.dropna().astype('float')

"""
#printing the sintering dataframe
print('Sintering Parameters: \n')
print(sintering_complete)
"""

#creating correlation used for heatmap
corr_sintering = sintering_complete.corr(method='pearson', numeric_only=True)

#sintering heatmap
plt.figure(figsize=(31, 22)) #changing size of heatmap
plt.title("Sintering Initial Heatmap", fontsize=40) #creating title for heatmap

#creating heatmap with seaborn
p2 = sns.heatmap(corr_sintering, cmap = 'summer', center = 0.2, linewidths = 7, annot = True, annot_kws={"size":40}, square = True,
                linecolor='black')
p2.set_xticklabels(p2.get_xticklabels(), rotation=20, fontsize=20)
p2.set_yticklabels(p2.get_yticklabels(), rotation=30, fontsize=21)

#saving sintering map as a png
sintering_path = *file path*
plt.savefig(sintering_path)
```

```python
#making a sintering heatmap with no empty spaces

#removing the unchanging rows from the heatmap
changing_vals_s = [0, 5, 6] 
sintering_final = sintering_complete.iloc[:, changing_vals_s]

#creating correlation used for heatmap
correlation_final_s = sintering_final.corr(method='pearson')

plt.figure(figsize=(15, 15)) #changing size of heatmap
plt.title("Sintering Final Heatmap", fontsize=40) #creating title for heatmap

#creating heatmap with seaborn
s1_final = sns.heatmap(correlation_final_s, cmap = 'Reds', linewidths = 3, annot = True, annot_kws={"size":25}, square = True)
s1_final.set_xticklabels(s1_final.get_xticklabels(), fontsize=15)
s1_final.set_yticklabels(s1_final.get_yticklabels(), fontsize=15)

#saving heatmap as a png image
sintering_finalpath = *file path*
plt.savefig(sintering_finalpath)
```
